```python

df['Defense/HP'] = df['Defense'] + df['HP']
df.head(5)
defense_hp_col = df['Defense/HP']
defense_col = df['Defense']
HP_col = df['HP']
def_hp_arr = defense_hp_col.to_numpy()
HP_arr = HP_col.to_numpy()
defense_arr = defense_col.to_numpy()
y_pred = np.array([])
for item in df['HP']:
    y_pred = np.append(y_pred, item)


y_pred.shape
plt.scatter(HP_arr, defense_arr)

plt.plot(HP_arr, y_pred, color='red', label ='predicted Line')
plt.title('Defense HP Relation for all pokemon')
plt.xlabel('HP')
plt.ylabel('Defense')
```
